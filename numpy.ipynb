```python
import numpy as np

a1 = np.array([1,2,3,4])
print(f"1D Array : {a1}")

a2 = np.array([
    [1,2,3],
    [4,5,6]
])
print(f"2D Array : {a2}")
```

```python
a3 = np.array([5,6,7,8])

print(f"{a1} + {a3} = {a1 + a3}")
```

```python
a4 = np.arange(10)
print(f"Array with arange : {a4}")
```
